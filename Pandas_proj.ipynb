```python
#lambda expressions can be used to generate new columns, pay attention the axis=1
def funs(x):
    return(x[0:2])
    
customers["initial"] = customers.apply(lambda y : funs(y["first_name"]),axis=1)
print(customers)
```

```python
sales = pd.read_csv("./sales.csv",parse_dates=["date"],dayfirst = True)
```

```python
pd.merge(customers,sales,left_on="customer_id",right_on="customer_id",how="inner")
```

```python
pd.merge(customers,sales,left_on="customer_id",right_on="customer_id",how="right")
```

```python
pd.merge(customers,sales,left_on="customer_id",right_on="customer_id",how="outer")
```

```python
data_merged = pd.merge(customers,sales,left_on="customer_id",right_on="customer_id",how="outer")
#group_by: aggregating data

data_merged.groupby(["customer_id","first_name"],as_index=False)["sales"].sum()

#we can get the sum,mean,max,etc.
```

```python
#Filtering data

#For example filter all the customers that have spent more than 80 on total

aggregated_sales_by_customer = data_merged.groupby(["customer_id","first_name"],as_index=False)["sales"].sum()

aggregated_sales_by_customer_filtered = aggregated_sales_by_customer[aggregated_sales_by_customer["sales"]>80]
print(aggregated_sales_by_customer_filtered)
```

```python
#sort a dataframe / let's sort by the customers who had the greatest sales
aggregated_sales_by_customer_filtered.sort_values(by="sales",ascending=False)
```

```python
#looping thru a dataframe
#For example, imagine those values are in dollars and we want to transform into british pounds

for i,row in aggregated_sales_by_customer_filtered.iterrows():
    print("i=" + str(i) + " / V=" + str(row["sales"] / 1.3))
```

```python
#selecting particular columns
print(aggregated_sales_by_customer_filtered[["customer_id","first_name"]])

#selecting particular rows

#For example element 0,0 (0=row,0=column)
print(aggregated_sales_by_customer_filtered.iat[0,0])

#this one works with indexes
print(aggregated_sales_by_customer_filtered.loc[0,"customer_id"])
```
